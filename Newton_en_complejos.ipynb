```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def polinomio(x):
  res = 1
  for raiz in lista:
    res *= (x-raiz)
  return res
```

```python
def derivada_polinomio(x):
  res = 0
  for cosa in lista:
    subres = 1
    copia = lista.copy()
    copia.remove(cosa)
    for raiz in copia:
      subres *= (x-raiz)
    res += subres
  return res
```

```python
def newton(x):
  for i in range(pasos_newton):
    u = derivada_polinomio(x)
    if(u !=0):
      x -= polinomio(x)/u
    else:
      return x
  return x

vec_newton = np.vectorize(newton)
```

```python
def colorear(x):
  arr = np.array(lista)
  arr = arr-x
  arr = np.multiply(arr, np.conj(arr)).real
  return np.argmin(arr)

vec_colorear = np.vectorize(colorear)
```

```python
def plot_campo():
  m,k = 2, 100
  x = np.linspace(-m, m, k)
  y = np.linspace(-m, m, k)
  xx, yy = np.meshgrid(x, y)
  zz = xx + 1j*yy
  zz = vec_newton(zz)

  z_R = zz.real
  z_I = zz.imag

  u_R = zz.real -xx
  u_I = zz.imag -yy


  fig = plt.figure(figsize=(10, 10))
  ax = fig.add_subplot(111)


  ax.quiver(xx, yy, u_R, u_I )
  color = 2 * np.log(np.hypot(z_R, z_I))
  ax.streamplot(x, y, z_R, z_I, color=color, linewidth=1, cmap=plt.cm.inferno, density=2, arrowstyle='->', arrowsize=1.5)

  ax.set_xlabel('Parte Real')
  ax.set_ylabel('Parte imaginaria')
  ax.set_xlim(-m,m)
  ax.set_ylim(-m,m)
  ax.set_aspect('equal')
  plt.show()

```

```python
b = math.cos((2/7)*math.pi) + math.sin((2/7)*math.pi)*1j
lista = [1,-1/2 + (3**(1/2)/2)*1j, -1/2 - (3**(1/2)/2)*1j]
pasos_newton = 5
plot_campo()
```

```python
def hacer_fractal(k):
  x = np.linspace(-2, 2, k)
  y = np.linspace(-2, 2, k)

  xx, yy = np.meshgrid(x, y)
  colores = xx + 1j*yy
  colores = vec_newton(colores)
  colores = vec_colorear(colores)

  plt_1 = plt.figure(figsize=(12, 12))
  h = plt.contourf(x, y, colores, cmap = 'tab20c')
  plt.axis('scaled')
  plt.colorbar()
  puntos = np.array(lista)
  plt.scatter(puntos.real,puntos.imag, color='red')
  plt.show()
```
